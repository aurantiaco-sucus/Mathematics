```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
x = np.linspace(-2,2,1000)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)
y5 = np.cbrt(x)
y6 = 1/x
y6[499] = np.nan
y_1 = np.repeat(1,1000)
plt.plot(x,y1,'r',x,y2,'g',x,y3,'b',x,y4,'y',x,y5,'m',x,y6,'c',x,y_1,'k',[1,1], [-2, 2],'k')
plt.axis('equal')
plt.ylim(-1,1)
plt.legend(['$y=x$','$y=x^2$','$y=x^3$','$y=\sqrt{x}$','$y=\sqrt[3]{x}$','$y=\\frac{1}{x}$'])
plt.grid()
plt.tight_layout()
plt.savefig('power.png', dpi=600)
```

```python
# plot y=2^x y=e^x
x = np.linspace(-2,2,1000)
y1 = 2**x
y2 = np.exp(x)
y_1 = np.repeat(1,1000)
plt.plot(x,y1,'r',x,y2,'g',x,y_1,'k')
plt.axis('equal')
plt.annotate('$(0,1)$', xy=(0, 1), xytext=(-2, 2), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('$(1,e)$', xy=(1, np.e), xytext=(2, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(['$y=2^x$','$y=e^x$'])
plt.ylim(0,4)
plt.grid()
plt.tight_layout()
plt.savefig('exponential.png', dpi=600)
```

```python
# plot y=ln(x)
x = np.linspace(0.01,4,1000)
y = np.log2(x)
y2 = np.log(x)
y3 = np.log10(x)
y_0 = np.repeat(0,1000)
plt.plot(x,y,'r',x,y2,'g',x,y3,'b',x,y_0,'k')
plt.axis('equal')
plt.ylim(-1.5,2)
plt.annotate('$(1,0)$', xy=(1, 0), xytext=(0.5, 0.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('$(e,1)$', xy=(np.e, 1), xytext=(2, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('$(\\frac{1}{e},-1)$', xy=(1/np.e, -1), xytext=(1, -1.25), arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(['$y=\log_2(x)$','$y=\ln(x)$','$y=\lg(x)$'])
plt.grid()
plt.tight_layout()
plt.savefig('logarithmic.png', dpi=600)
```

```python
# plot y=sin(x), y=cos(x), y=tan(x)
x = np.linspace(-2*np.pi,2*np.pi,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y3[y3 > 16] = np.nan
y3[y3 < -16] = np.nan
y_0 = np.repeat(0,1000)
y_1 = np.repeat(1,1000)
y_sqrt2div2 = np.repeat(np.sqrt(2)/2,1000)
plt.plot(x,y1,'r',x,y2,'g',x,y3,'b',x,y_0,'k',x,y_1,'k',[0,0], [-16, 16],'k',x,y_sqrt2div2,'y')
plt.axis('equal')
plt.ylim(-1,1)
plt.xticks([-2*np.pi, -1.75*np.pi, -1.5*np.pi, -1.25*np.pi, -1*np.pi, -0.75*np.pi, -0.5*np.pi, -0.25*np.pi, 0, 0.25*np.pi, 0.5*np.pi, 0.75*np.pi, 1*np.pi, 1.25*np.pi, 1.5*np.pi, 1.75*np.pi, 2*np.pi], ['$-2\pi$', '$-\\frac{7}{4}\pi$', '$-\\frac{3}{2}\pi$', '$-\\frac{5}{4}\pi$', '$-\pi$', '$-\\frac{3}{4}\pi$', '$-\\frac{1}{2}\pi$', '$-\\frac{1}{4}\pi$', '$0$', '$\\frac{1}{4}\pi$', '$\\frac{1}{2}\pi$', '$\\frac{3}{4}\pi$', '$\\pi$', '$\\frac{5}{4}\pi$', '$\\frac{3}{2}\pi$', '$\\frac{7}{4}\pi$', '$2\pi$'])
plt.grid()
plt.annotate('$(\\frac{\pi}{4},\\frac{\sqrt{2}}{2})$', xy=(np.pi/4, np.sqrt(2)/2), xytext=(0.5, -3), arrowprops=dict(facecolor='black', shrink=0.05))
plt.tight_layout()
plt.savefig('trigonometric1.png', dpi=600)
```

```python
# plot cot(x), sec(x), csc(x)
x = np.linspace(-1*np.pi,1*np.pi,1000)
y1 = 1/np.tan(x)
y1[y1 > 16] = np.nan
y1[y1 < -16] = np.nan
y2 = 1/np.cos(x)
y2[249] = np.nan
y2[749] = np.nan
y3 = 1/np.sin(x)
y3[y3 > 16] = np.nan
y3[y3 < -16] = np.nan
y_0 = np.repeat(0,1000)
y_1 = np.repeat(1,1000)
plt.plot(x,y1,'r',x,y2,'g',x,y3,'b',x,y_0,'k',x,y_1,'k',[0,0], [-16, 16],'k')
plt.axis('equal')
plt.ylim(-1,1)
plt.xticks([-1*np.pi, -0.75*np.pi, -0.5*np.pi, -0.25*np.pi, 0, 0.25*np.pi, 0.5*np.pi, 0.75*np.pi, 1*np.pi], ['$-\pi$', '$-\\frac{3}{4}\pi$', '$-\\frac{1}{2}\pi$', '$-\\frac{1}{4}\pi$', '$0$', '$\\frac{1}{4}\pi$', '$\\frac{1}{2}\pi$', '$\\frac{3}{4}\pi$', '$\pi$'])
plt.grid()
plt.legend(['$y=\cot(x)$','$y=\sec(x)$','$y=\csc(x)$'])
plt.tight_layout()
plt.savefig('trigonometric2.png', dpi=600)
```

```python
# plot y=arcsin(x), y=arccos(x), y=arctan(x)
x = np.linspace(-1,1,1000)
y1 = np.arcsin(x)
y2 = np.arccos(x)
x2 = np.linspace(-10,10,10000)
y3 = np.arctan(x2)
plt.plot(x,y1,'r',x,y2,'g',x2,y3,'b')
plt.axis('equal')
plt.ylim(-np.pi,np.pi)
plt.xlim(-1,1)
plt.legend(['$y=\\arcsin(x)$','$y=\\arccos(x)$','$y=\\arctan(x)$'])
plt.yticks([-1*np.pi, -0.75*np.pi, -0.5*np.pi, -0.25*np.pi, 0, 0.25*np.pi, 0.5*np.pi, 0.75*np.pi, 1*np.pi], ['$-\pi$', '$-\\frac{3}{4}\pi$', '$-\\frac{1}{2}\pi$', '$-\\frac{1}{4}\pi$', '$0$', '$\\frac{1}{4}\pi$', '$\\frac{1}{2}\pi$', '$\\frac{3}{4}\pi$', '$\pi$'])
plt.grid()
plt.tight_layout()
plt.savefig('inverse trigonometric.png', dpi=600)
```
